```python
#4
import nltk
from nltk.corpus import stopwords

# Tải bổ sung các gói cần thiết cho việc tách từ và lọc
nltk.download('punkt_tab')
nltk.download('punkt')
nltk.download('stopwords')

# Code xử lý Câu 4
text = "This is a simple example to demonstrate the removal of stopwords from a text."
words = nltk.word_tokenize(text)
stop_words_eng = set(stopwords.words('english'))

filtered_words = [w for w in words if w.lower() not in stop_words_eng]
print("Văn bản ban đầu:", words)
print("Sau khi loại bỏ stopwords:", filtered_words)
```

```python
#5
custom_stopwords = set(stopwords.words('english'))
# Bỏ qua từ 'not' và 'no' khỏi danh sách stopwords
custom_stopwords.difference_update({'not', 'no'})
print("Có chữ 'not' trong stopwords không?", 'not' in custom_stopwords)
```

```python
#6
syns = wordnet.synsets("program")
# Lấy nghĩa đầu tiên
print("Nghĩa của từ:", syns[0].definition())
print("Ví dụ:", syns[0].examples())
```

```python
#7
synonyms = []
antonyms = []

for syn in wordnet.synsets("good"):
    for l in syn.lemmas():
        synonyms.append(l.name())
        if l.antonyms():
            antonyms.append(l.antonyms()[0].name())

print("Đồng nghĩa của 'good':", set(synonyms))
print("Trái nghĩa của 'good':", set(antonyms))
```

```python
#9
# Tính độ tương đồng giữa "dog" và "cat"
dog = wordnet.synset('dog.n.01')
cat = wordnet.synset('cat.n.01')
print("Độ tương đồng giữa dog và cat:", dog.path_similarity(cat))
```

```python
#10
run = wordnet.synset('run.v.01')
walk = wordnet.synset('walk.v.01')
print("Độ tương đồng giữa run và walk:", run.path_similarity(walk))
```

```python
#11
male_names = names.words('male.txt')
female_names = names.words('female.txt')

print(f"Số lượng tên nam: {len(male_names)}")
print("10 tên nam đầu tiên:", male_names[:10])

print(f"\nSố lượng tên nữ: {len(female_names)}")
print("10 tên nữ đầu tiên:", female_names[:10])
```

```python
#12
# Gắn nhãn cho tên
labeled_names = ([(name, 'male') for name in male_names] +
                 [(name, 'female') for name in female_names])

# Trộn ngẫu nhiên
random.shuffle(labeled_names)
print("15 kết hợp ngẫu nhiên đầu tiên:\n", labeled_names[:15])
```

```python
#13
# Tạo mảng chứa ký tự cuối và nhãn giới tính tương ứng
featuresets = [(name[-1], gender) for (name, gender) in labeled_names]
print("Mảng đặc trưng (ký tự cuối, giới tính) - 10 mẫu đầu:\n", featuresets[:10])
```
